## What this notebook demonstrates

PatientDoctorBridge breaks the language barrier between non-English-speaking patients and doctors in Indian clinics. It runs **100% locally** — no cloud calls, no data leaves the device.

| Scenario | Description | Gemma 4 model |
|---|---|---|
| **Bridge Translation** | Patient speech (Hindi/Telugu/Kannada/Tamil) ↔ Doctor English | `gemma4:e2b` |
| **Emergency Triage** | Structured JSON extraction from patient speech | `gemma4:e4b` |
| **Prescription OCR** | Handwritten prescription → structured medicine list | `gemma4:e4b` (vision) |
| **Emergency Reassurance** | Fast English phrase → patient language translation | `gemma4:e2b` |

### Architecture
```
Patient speaks → [ASR: Whisper or IndicConformer] → transcript → [Gemma 4] → Doctor English
Doctor speaks  → [ASR: Whisper]                   → transcript → [Gemma 4] → Patient language
Doctor reply   → [MMS-TTS]                        → audio read-aloud in patient language
Prescription image                                → [Gemma 4 vision] → JSON → [MMS-TTS] → read-aloud
```

All LLM inference runs via **Ollama** using:
- `gemma4:e2b` — FAST_TRANSLATION mode (Patient↔Doctor, Reassurance, Prescription summary)
- `gemma4:e4b` — REASONING_EXTRACTION mode (Triage, OCR)

### New features since initial release
- **ASR backend option:** IndicConformer-600M (AI4Bharat) for better Indian language transcription — selectable per-request in the web UI
- **TTS (Text-to-Speech):** Facebook MMS-TTS for patient-language read-aloud of translations, reassurance phrases, and prescription summaries
- **Audio preprocessing pipeline:** SNR analysis, silence detection, and light noise gate before ASR
- **Script normalization for Hindi:** Urdu/romanized script detection and automatic Devanagari correction
- **Whisper hallucination guard:** Detects and breaks repeat-token loops in transcription output
- **Retry logic with exponential backoff:** Automatic retries for timeout/connection errors to Ollama
- **Increased timeouts:** `OLLAMA_TIMEOUT`: 180s (fast translation), `OLLAMA_TIMEOUT_REASONING`: 900s (triage/OCR)

````python
# ═══════════════════════════════════════════════════════════════
# STEP 6 — Pull Gemma 4 Models
# Competition rule: ONLY gemma4:e2b and gemma4:e4b are permitted.
# ═══════════════════════════════════════════════════════════════
import subprocess
import time

REQUIRED_MODELS = ["gemma4:e2b", "gemma4:e4b"]

def model_available(tag: str) -> bool:
    r = subprocess.run(["ollama", "list"], capture_output=True, text=True)
    return tag in r.stdout

def pull_model(tag: str):
    if model_available(tag):
        print(f"✅ {tag} already available (skipping pull)")
        return
    print(f"⬇️  Pulling {tag}…  (this may take several minutes on first run)")
    r = subprocess.run(["ollama", "pull", tag], capture_output=True, text=True)
    if r.returncode == 0:
        print(f"✅ {tag} ready")
    else:
        print(f"❌ Failed to pull {tag}")
        print(r.stderr[-500:])

for model in REQUIRED_MODELS:
    pull_model(model)

print()
print("Installed models:")
subprocess.run(["ollama", "list"])
````

## 🔊 Audio Pipeline Features

PatientDoctorBridge includes a full audio processing pipeline beyond basic ASR:

### Audio Preprocessing (`audio/preprocessor.py`)
- **SNR (Signal-to-Noise Ratio) analysis** — measures recording quality before transcription
- **Silence detection** — identifies recordings with insufficient speech content
- **Light noise gate** — attenuates low-level background noise for cleaner ASR input

### Script Normalization (`audio/script_normalizer.py`)
- **Hindi script recovery** — detects Urdu-script or romanized-Latin Hindi and converts to Devanagari
- Ensures Whisper output is in the expected script before passing to Gemma 4 for translation

### Whisper Hallucination Guard (`audio/handler.py`)
- Detects repeat-token loops (a known Whisper failure mode on short/noisy audio)
- Breaks the loop early and returns the valid prefix, preventing garbage output

### Optional: IndicConformer-600M ASR (`audio/indic_conformer.py`)
- AI4Bharat's `indic-conformer-600m-multilingual` model
- Produces native-script output (Devanagari / Telugu / Kannada / Tamil) on the first pass
- Selectable per-request in the web UI via a "Speech model" dropdown
- Hybrid auto-detect mode: Whisper detects language → IndicConformer re-transcribes with locked language

### Text-to-Speech: MMS-TTS (`audio/tts.py`)
- Facebook's Massively Multilingual Speech TTS (local VITS model)
- Provides patient-language read-aloud for Doctor→Patient translations, reassurance phrases, and prescription summaries
- No OS voice pack or cloud TTS API required — fully local inference

## 💊 Scenario 3: Emergency Reassurance Phrases

````python
# ── Scenario 3: Emergency Reassurance — All Languages ────────────────────────
from rich.table import Table

from translation.reassurance import REASSURANCE_PHRASES, ReassuranceService

reassure_svc = ReassuranceService(engine)

TARGET_LANGUAGES = ["hi", "te", "kn", "ta"]
LANG_NAMES = {"hi": "Hindi", "te": "Telugu", "kn": "Kannada", "ta": "Tamil"}

# Test all URGENT phrases across all 4 languages
urgent_phrases = [p for cat, p in REASSURANCE_PHRASES if cat == "URGENT"]

for lang_code in TARGET_LANGUAGES:
    t = Table(
        title=f"Reassurance → {LANG_NAMES[lang_code]}",
        show_header=True,
        header_style="bold green",
    )
    t.add_column("English (Original)", style="dim", min_width=40)
    t.add_column(f"{LANG_NAMES[lang_code]} Translation")

    for phrase in urgent_phrases:
        translated = reassure_svc.translate(phrase, lang_code)
        t.add_row(phrase, translated)

    console.print(t)
    print()
````

## 📋 Scenario 4: Prescription OCR + Patient-Language Translation

Patients often can't read their own prescriptions (handwritten or in English). Gemma 4 (`gemma4:e4b`, multimodal vision) reads the prescription image and extracts a structured list of medicines with dosage, frequency, and instructions.

**New (Phase 4b):** After OCR, Gemma 4 (`gemma4:e2b`) translates the extracted prescription into the patient's native language as a plain, easy-to-understand summary.

**Language selector:** Set `PRESCRIPTION_LANG` below before running.


````python
# ── Scenario 4: Prescription OCR + Translation ───────────────────────────────
import os

from rich.table import Table

from translation.prescription import PrescriptionService

rx_svc = PrescriptionService(engine)

# ── Language selector ────────────────────────────────────────────────────────
# Set the target language for the patient-friendly prescription summary.
# Change PRESCRIPTION_LANG to any of the supported codes:
#   "hi" = Hindi | "te" = Telugu | "kn" = Kannada | "ta" = Tamil | "en" = English (no translation)
PRESCRIPTION_LANG = "hi"  # ← change this before running

LANG_NAMES = {"hi": "Hindi", "te": "Telugu", "kn": "Kannada", "ta": "Tamil", "en": "English"}
lang_display = LANG_NAMES.get(PRESCRIPTION_LANG, PRESCRIPTION_LANG)

if not SAMPLE_IMAGES:
    print("⚠️  No sample prescription images found in project root.")
    print("   Expected: Prescription1.png, Prescription2.png, Prescription3.png")
else:
    for img_path in SAMPLE_IMAGES:
        img_name = os.path.basename(img_path)
        print(f"\n{'━'*60}")
        print(f"Processing: {img_name}")
        if PRESCRIPTION_LANG != "en":
            print(f"Translation target: {lang_display}")
        print(f"{'━'*60}")

        # Display the prescription image
        from IPython.display import Image as IPyImage
        from IPython.display import display
        display(IPyImage(img_path, width=500))

        try:
            # ── Step 1: OCR extraction (gemma4:e4b vision) ───────────────────
            result = rx_svc.extract(img_path)

            # Meta info
            meta_table = Table(title=f"Prescription Metadata — {img_name}",
                               show_header=False)
            meta_table.add_column("Field", style="dim")
            meta_table.add_column("Value")
            meta_table.add_row("Doctor",  result["doctor_name"])
            meta_table.add_row("Patient", result["patient_name"])
            meta_table.add_row("Date",    result["date"])
            meta_table.add_row("Notes",   result["notes"] or "—")
            console.print(meta_table)

            # Medicines table (English, extracted by OCR)
            if result["medicines"]:
                med_table = Table(title="Medicines Extracted (English)",
                                  show_header=True, header_style="bold purple")
                med_table.add_column("#", min_width=2)
                med_table.add_column("Medicine", min_width=16)
                med_table.add_column("Dosage")
                med_table.add_column("Form")
                med_table.add_column("Frequency")
                med_table.add_column("Duration")
                med_table.add_column("Instructions")

                for i, m in enumerate(result["medicines"], 1):
                    med_table.add_row(
                        str(i), m["name"], m["dosage"], m["form"],
                        m["frequency"], m["duration"], m["instructions"],
                    )
                console.print(med_table)

                # ── Step 2: Translate to patient language (gemma4:e2b) ───────
                if PRESCRIPTION_LANG != "en":
                    print(f"\nTranslating prescription summary to {lang_display}…")
                    translated = rx_svc.translate_summary(result, PRESCRIPTION_LANG)

                    trans_table = Table(
                        title=f"Patient Summary in {lang_display} (for patient to read)",
                        show_header=False,
                        header_style="bold green",
                    )
                    trans_table.add_column("Summary", min_width=60, style="green")
                    trans_table.add_row(translated)
                    console.print(trans_table)
                else:
                    print("ℹ️  PRESCRIPTION_LANG='en' — no translation performed.")
                    print("   Change PRESCRIPTION_LANG to 'hi', 'te', 'kn', or 'ta' above.")
            else:
                print("⚠️  No medicines extracted from this image.")

        except Exception as exc:
            print(f"❌ Failed for {img_name}: {exc}")
````

### Key technical highlights
- **Hardware-adaptive configuration**: auto-selects GPU float16 or CPU int8 for Whisper; Ollama picks optimal quantization per device
- **Privacy-first**: all inference local via Ollama; audio deleted immediately after transcription
- **Gemma 4 only**: `gemma4:e2b` for speed, `gemma4:e4b` for reasoning/vision — no other LLMs used
- **Resilient JSON parsing**: 4-strategy fallback handles truncated Gemma 4 output
- **Retry logic with exponential backoff**: automatic retries on timeout/connection errors to Ollama
- **Audio preprocessing & script normalization**: SNR analysis, silence detection, Hindi Devanagari recovery from romanized/Urdu script
- **Whisper hallucination guard**: repeat-token loop detection prevents garbage ASR output
- **TTS integration**: Facebook MMS-TTS for local patient-language read-aloud (no cloud APIs)
- **IndicConformer ASR backend option**: AI4Bharat's 600M model for superior Indian-language transcription, selectable per-request
- **Production-grade web UI**: WebM recording, language auto-detection, TTS auto-playback, print-friendly prescription view
